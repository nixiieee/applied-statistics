```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Практика

```python
def generate_uniform(N, a, b):
    return np.random.uniform(a,b,N)
```

```python
np.mean(generate_uniform(100, 0, 1))
```

```python
plt.hist(generate_uniform(10000, 0, 1), bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Гистограмма равномерного распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.show()
```

```python
np.mean(generate_uniform(100, -1, 1))
```

```python
plt.hist(generate_uniform(10000, -1, 1), bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Гистограмма равномерного распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')
plt.show()
```

```python
plt.hist(generate_uniform(10000, -1, 1) + generate_uniform(10000, -1, 1) + generate_uniform(10000, -1, 1), bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.show()
```

```python
def generate_cauchy(N):
    return np.random.standard_cauchy(N)
```

```python
plt.hist(generate_cauchy(10000), bins=10000, density=True, alpha=0.7, color='blue', edgecolor='black')
#plt.xlim(-25, 25)
plt.show()
```

```python
plt.hist(generate_cauchy(10000), bins=10000, density=True, alpha=0.7, color='blue', edgecolor='black')
plt.xlim(-25, 25)
plt.show()
```

```python
def to_fixed_point(value, bits=16):
    max_value = 2 ** (bits - 1)
    int_representation = int((value + 1) * max_value / 2)  # Масштабирование на отрезок [0, 2^bits-1]
    return f'{int_representation:0{bits}b}'

def convert_to_binary_string(numbers):
    binary_string = ''.join([to_fixed_point(num) for num in numbers])
    return binary_string
```

```python
binary_string = convert_to_binary_string(generate_uniform(100, 0, 1))
with open("binary_string.txt", "w") as f:
    f.write(binary_string)
```
